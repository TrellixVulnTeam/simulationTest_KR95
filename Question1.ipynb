```python
bnx = gum.BayesNet('x')
t = bnx.add(gum.LabelizedVariable('t', 'tired', 2))
f = bnx.add(gum.LabelizedVariable('f', 'fall_asleep', 2))
bnx.addArc(t, f)
bnx.cpt(t).fillWith([0.75, 0.25])# P(tired) = 0.2
bnx.cpt(f)[{'t':0}] = [0.95, 0.05]
bnx.cpt(f)[{'t':1}] = [0.2, 0.8]
ie=gum.LazyPropagation(bnx)
# if E = 0, what's H = 1? -> 0.06
(0.2*0.25)/(0.2*0.25+0.95*0.75) # 0.06557377049180328
gnb.showInference(bnx,evs={'f':0})
```

```python
bny = gum.BayesNet('y')
a = bny.add(gum.LabelizedVariable('a', 'a', 2))
b = bny.add(gum.LabelizedVariable('b', 'b', 2))
c = bny.add(gum.LabelizedVariable('c', 'c', 2))

bny.addArc(a,c)
bny.addArc(b,c)


bny.cpt(a).fillWith([0.75, 0.25])# P(tired) = 0.2
bny.cpt(b).fillWith([0.01, 0.99])# P(tired) = 0.2


bny.cpt(c)[{'a':0, 'b':0}] = [0.95, 0.05]
bny.cpt(c)[{'a':1, 'b':0}] = [0.2, 0.8]
bny.cpt(c)[{'a':0, 'b':1}] = [0.95, 0.05]
bny.cpt(c)[{'a':1, 'b':1}] = [0.2, 0.8]
ie=gum.LazyPropagation(bny)
gnb.showInference(bny,evs={})
```
